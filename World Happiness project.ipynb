```python
#Transforming the Data
```

```python
df = df[['Country', 'Region', 
        'Happiness Score',
        'Economy (GDP per Capita)', 'Family',
       'Health (Life Expectancy)', 'Freedom', 'Corruption',
       'Generosity', 
        ]].copy()
df.head()
```

```python
df = df.rename(columns={'Trust (Government Corruption)':'Corruption'})
df
```

```python
#Analysis of Data and Visualization
```

```python
reagion_avg = df.groupby(['Region']).mean(numeric_only=True).sort_values('Happiness Score', ascending=False)
reagion_avg
```

```python
#Visualize the above table
```

```python
ax1 = reagion_avg.plot( y=['Economy (GDP per Capita)','Family', 'Health (Life Expectancy)',
                           'Freedom','Corruption','Generosity'],
                       figsize=(15,5), kind='bar', rot=45)
ax1.set_ylabel('Score')
ax1.set_title('Average Scores by Region')
plt.show()
```
